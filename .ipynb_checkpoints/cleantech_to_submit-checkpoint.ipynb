```python
## column_header from the sample(Gen1) folder
data_sample_18_i1 = pd.read_csv("./Gen1/[IN-023C]/2018/2018-12/Inverter_1/[IN-023C]-I1-2018-12-01.txt",sep='\t')
data_sample_18_i2 = pd.read_csv("./Gen1/[IN-023C]/2018/2018-12/Inverter_2/[IN-023C]-I2-2018-12-01.txt",sep='\t')
data_sample_18_mfm = pd.read_csv('./Gen1/[IN-023C]/2018/2018-12/MFM/[IN-023C]-MFM1-2018-12-01.txt',sep='\t')
data_sample_18_wms = pd.read_csv('./Gen1/[IN-023C]/2018/2018-12/WMS/[IN-023C]-WMS-2018-12-01.txt',sep='\t')

# data_sample_19_i1 = pd.read_csv('./Gen1/[IN-023C]/2019/2019-01/Inverter_1/[IN-023C]-I1-2019-01-01.txt',sep='\t')
# data_sample_19_i2 = pd.read_csv('./Gen1/[IN-023C]/2019/2019-01/Inverter_2/[IN-023C]-I2-2019-01-01.txt',sep='\t')
# data_sample_19_mfm = pd.read_csv('./Gen1/[IN-023C]/2019/2019-01/MFM/[IN-023C]-MFM1-2019-01-01.txt',sep='\t')
# data_sample_19_wms = pd.read_csv('./Gen1/[IN-023C]/2019/2019-01/WMS/[IN-023C]-WMS-2019-01-01.txt',sep='\t')
    
colheader_sample_i1 = list(data_sample_18_i1.columns)
colheader_sample_i2 = list(data_sample_18_i2.columns)
colheader_sample_mfm = list(data_sample_18_mfm.columns)
colheader_sample_wms = list(data_sample_18_wms.columns)

# colheader_sample_19_i1 = list(data_sample_19_i1.columns)
# colheader_sample_19_i2 = list(data_sample_19_i2.columns)
# colheader_sample_19_mfm = list(data_sample_19_mfm.columns)
# colheader_sample_19_wms = list(data_sample_19_wms.columns)
```

```python
# iterate path variable for filenames

list_file_raw_18_i1 =[file for file in sorted(os.listdir(pathvar_raw_18_i1)) if '.txt' in file]
list_file_raw_18_i2 =[file for file in sorted(os.listdir(pathvar_raw_18_i2)) if '.txt' in file]
list_file_raw_18_mfm =[file for file in sorted(os.listdir(pathvar_raw_18_mfm)) if '.txt' in file]
list_file_raw_18_wms =[file for file in sorted(os.listdir(pathvar_raw_18_wms)) if '.txt' in file]

list_file_raw_19_i1 =[file for file in sorted(os.listdir(pathvar_raw_19_i1)) if '.txt' in file]
list_file_raw_19_i2 =[file for file in sorted(os.listdir(pathvar_raw_19_i2)) if '.txt' in file]
list_file_raw_19_mfm =[file for file in sorted(os.listdir(pathvar_raw_19_mfm)) if '.txt' in file]
list_file_raw_19_wms =[file for file in sorted(os.listdir(pathvar_raw_19_wms)) if '.txt' in file]
```

```python
def col_swap(data,string1,string2):
    cols = list(data.columns)
    a, b = cols.index(string1), cols.index(string2)
    cols[b], cols[a] = cols[a], cols[b]
    data = data[cols]
    raw_data_columns = list(data.columns)
    return raw_data_columns
```

```python
def calculate_year_18(list_file, string):
    for i in list_file:
        if i.split('-')[2]==string:
            
            if i.split('-')[2]=='I1':
                
                data = pd.read_csv(pathvar_raw_18_i1/i,sep = '\t')
                
                raw_data_columns= col_swap(data,'i1','i32')
            
                d = dict(zip(raw_data_columns,colheader_sample_i1))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                data.iloc[:,[1,3]]=data.iloc[:,[1,3]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_i1}/{i}",sep='\t',index=False)
                
                print(f"success in {i}")
            
            elif i.split('-')[2]=='I2':
                
                data = pd.read_csv(pathvar_raw_18_i2/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'i1','i32')
            
                d = dict(zip(raw_data_columns,colheader_sample_i1))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                data.iloc[:,[1,3]]=data.iloc[:,[1,3]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_i2}/{i}",sep='\t',index=False)
                
                print(f"success in {i}")
                
            elif i.split('-')[2]=='MFM1':
                
                data = pd.read_csv(pathvar_raw_18_mfm/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'m1','m63')
            
                d = dict(zip(raw_data_columns,colheader_sample_mfm))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_mfm}/{i}",sep='\t',index=False)
                
                print(f"success in {i}")
                
            elif i.split('-')[2]=='WMS':
                
                data = pd.read_csv(pathvar_raw_18_wms/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'w1','w23')
            
                d = dict(zip(raw_data_columns,colheader_sample_wms))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                if i.split('-')[-1].split('.')[0]!='30':
                    data.iloc[:,7:]=data.iloc[:,7:].astype('float64')
                else:
                    data.iloc[:,8:]=data.iloc[:,8:].astype('float64')
                data.iloc[:,[1,3,5]]=data.iloc[:,[1,3,5]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_wms}/{i}",sep='\t',index=False)
            
                print(f"success in {i}")
```
