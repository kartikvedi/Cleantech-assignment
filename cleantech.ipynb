```python
def col_swap(data,string1,string2):
    cols = list(data.columns)
    a, b = cols.index(string1), cols.index(string2)
    cols[b], cols[a] = cols[a], cols[b]
    data = data[cols]
    raw_data_columns = list(data.columns)
    return raw_data_columns
```

```python
def calculate_year_18(list_file, string):
    for i in list_file:
        if i.split('-')[2]==string:
            
            if i.split('-')[2]=='I1':
                print("entered inverter1")
                print(i.split('-')[2])
                data = pd.read_csv(pathvar_raw_18_i1/i,sep = '\t')
                
                raw_data_columns= col_swap(data,'i1','i32')
            
                d = dict(zip(raw_data_columns,colheader_sample_i1))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                data.iloc[:,[1,3]]=data.iloc[:,[1,3]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_i1}/{i}",sep='\t',index=False)
            
            elif i.split('-')[2]=='I2':
                print("entered inverter2")
                
                print(i.split('-')[2])
                data = pd.read_csv(pathvar_raw_18_i2/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'i1','i32')
            
                d = dict(zip(raw_data_columns,colheader_sample_i1))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                data.iloc[:,[1,3]]=data.iloc[:,[1,3]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_i2}/{i}",sep='\t',index=False)
                
            elif i.split('-')[2]=='MFM1':
                print("entered mfm")
                
                print(i.split('-')[2])
                data = pd.read_csv(pathvar_raw_18_mfm/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'m1','m63')
            
                d = dict(zip(raw_data_columns,colheader_sample_mfm))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                data.iloc[:,5:]=data.iloc[:,5:].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_mfm}/{i}",sep='\t',index=False)
                
            elif i.split('-')[2]=='WMS':
                print("entered wms")
                
                print(i.split('-')[2])
                data = pd.read_csv(pathvar_raw_18_wms/i,sep = '\t')
            
                raw_data_columns= col_swap(data,'w1','w23')
            
                d = dict(zip(raw_data_columns,colheader_sample_wms))
                data = data[d]
                data.rename(columns = d,inplace = True)
                
                if i.split('-')[-1].split('.')[0]!='30':
                    data.iloc[:,7:]=data.iloc[:,7:].astype('float64')
                else:
                    data.iloc[:,8:]=data.iloc[:,8:].astype('float64')
                data.iloc[:,[1,3,5]]=data.iloc[:,[1,3,5]].astype('float64')
                
                data.fillna('NULL',inplace=True)
                data.to_csv(f"{pathvar_gen_18_wms}/{i}",sep='\t',index=False)
            
                print(f"success in {i}")
```
